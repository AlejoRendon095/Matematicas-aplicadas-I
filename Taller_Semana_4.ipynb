**Serie Fibonacci:** Escribe un programa en Python que genere los primeros n números de la serie de Fibonacci y visualízalos en un gráfico de líneas utilizando Matplotlib. El usuario debe poder elegir el número n


```python
import numpy as np
import matplotlib.pyplot as plt

print("Serie Fibonacci")
n = int(input("Ingrese el numero de elementos de la serie: "))

def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

print("Los primeros", n, "elementos de la serie de Fibonacci son:", fibonacci(n))

x = np.linspace(0,n-1, n)
y = fibonacci(n)

plt.plot(x, y, ".", ms = 2, color = "red")
plt.title("Serie Fibonacci")
plt.grid(linestyle = "--")
plt.show()
```

**Análisis de datos meteorológicos:** Dado un conjunto de datos con las temperaturas

**temperaturas** = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

```python
import numpy as np
import matplotlib.pyplot as plt

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = sum(temperaturas)/len(temperaturas)
print ("La media de las temperaturas proporcionadas es de: " +str(media))

mediana = np.median(temperaturas)
print ("La mediana de la lista de temperaturas proporcionadas es de: " +str(mediana))

desviacion_estandar = np.sqrt(sum((np.array(temperaturas) - media)**2)/len(temperaturas))
print ("La desviacion estandar de la lista de temperaturas proporcionadas es de: " +str(desviacion_estandar))


y = temperaturas
x = np.linspace(1,31,20)

plt.ylabel ("Temperaturas en C°")
plt.xlabel ("Dias")
plt.title ("Temperaturas mensual")
plt.grid(linestyle = "--")
plt.plot(x, y, "o:y", ms = 10, color = "red")
plt.show()


```

**Simulación de un dado:** Simula 1000 lanzamientos de un dado tradicional y muestra la frecuencia de cada número en un gráfico de barras


```python
import matplotlib.pyplot as plt
import random as rd
import numpy as np

def dado():
  return rd.randint(1,6)

dadoSim = []
for i in range(1000):
  dadoSim.append(dado())



frecuencias = [dadoSim.count(i) for i in range(1, 7)]


plt.bar(range(1, 7), frecuencias, color='skyblue', edgecolor='black')
plt.xlabel("Número en el dado")
plt.ylabel("Frecuencia")
plt.title("Frecuencia de cada número en 1000 lanzamientos de un dado")
plt.grid(linestyle='--', alpha=0.7)
plt.xticks(range(1, 7))
plt.show()


print(dadoSim)
print(len(dadoSim))
```

**Gráfica de funciones trigonométricas:** Escribe un programa que grafique las
funciones seno, coseno y tangente en el mismo plano para valores entre 0 y 2π.

```python
import numpy as np
import matplotlib.pyplot as plt

def seno(x):
  return np.sin(x)

def coseno(x):
  return np.cos(x)

def tangente(x):
  return np.tan(x)

x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, seno(x), label = "seno")
plt.plot(x, coseno(x), label = "coseno")
plt.plot(x, tangente(x), label = "tangente")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linestyle = "--")
plt.title("Funciones trigonométricas")
plt.xlim(0, 2*np.pi)
plt.ylim(-2, 2)
plt.axvline(x=0, color='black', linestyle='-')
plt.axhline(y=0, color='black')
plt.show()
```

**Movimiento parabólico:** Escribe un programa que simule el movimiento de un
proyectil bajo la influencia de la gravedad (sin resistencia del aire). Grafica la trayectoria parabólica para diferentes ángulos de lanzamiento

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd


g = 9.81

# @title Simulador del proyectil { run: "auto", vertical-output: true }
Velocidad_inicial = 80 # @param {type:"number"}
Angulo_de_lanzamiento = 45 # @param {type:"number"}

Angulo_de_lanzamiento = np.radians(Angulo_de_lanzamiento)

def vuelo():
  total_vuelo = 2*Velocidad_inicial*np.sin(Angulo_de_lanzamiento)/g
  return total_vuelo


tiempo = np.linspace(0,vuelo(),100)
x = Velocidad_inicial*np.cos(Angulo_de_lanzamiento)*tiempo
y = Velocidad_inicial*np.sin(Angulo_de_lanzamiento)*tiempo - 0.5*g*tiempo**2

plt.plot(x,y, "-", ms = 1)
plt.ylabel("Altura en metros")
plt.xlabel("Alcance en metros")
plt.grid(linestyle = "--")
plt.title("Trayectoria del proyectil")
plt.pause(0.1)

plt.show()



```

Punto Extra: **Fractales con Python:** Escribe un programa que genere el fractal de Sierpinski utilizando ciclos (while), y visualiza el resultado con Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

def midpoint(p1, p2):
    return [(p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2]

vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])

p = [0.5, 0.25]

points = [p]

n = 50000
i = 0

while i < n:

    vertex = vertices[np.random.randint(0, 3)]
    p = midpoint(p, vertex)
    points.append(p)
    i += 1

points = np.array(points)

plt.figure(figsize=(8, 8))
plt.scatter(points[:, 0], points[:, 1], s=0.1, color='blue')
plt.title("Fractal de Sierpinski")
plt.axis('equal')
plt.axis('off')
plt.show()
```
