#<font color = "Blue">$[3]$ $Metodo$ $De$ $Newton$

Para cada una de las siguientes ecuaciones, utilice un gráfico para obtener estimaciones iniciales de las soluciones y luego aplique el método de Newton para localizar las soluciones dentro de 0,0001. Utilice el código

$a)$  $x^5-6x^3+2x = 2$

```python
import numpy as np
import matplotlib.pyplot as plt

def func_A(x):
    return x**5 - 6*x**3 + 2*x - 2

def dfunc_A(x):
    return 5*x**4 - 18*x**2 + 2

def newton_method(func, dfunc, x0, tolerance=0.0001, max_iter=100):
    x = x0
    for _ in range(max_iter):
        fx = func(x)
        if abs(fx) < tolerance:
            return x
        dfx = dfunc(x)
        if dfx == 0:
            break
        x = x - fx / dfx
    return None

x = np.linspace(-2, 2.5, 500)
y = func_A(x)
plt.plot(x, y)
plt.axhline(0, color='black', linestyle='-')
plt.axvline(0, color='black', linestyle='-')
plt.title("f(x) = x^5 - 6x^3 + 2x - 2")
plt.xlabel("x")
plt.ylabel("f(x)")

initial_guesses = [-1.5, -0.5, 0.5, 1.5]
intersections_x = []

solutions = []
for guess in initial_guesses:
    sol = newton_method(func_A, dfunc_A, guess)
    if sol is not None and all(abs(sol - s) > 0.0001 for s in solutions):
        solutions.append(sol)
        intersections_x.append(sol)

plt.scatter(intersections_x, [0]*len(intersections_x), color="red", zorder=5, label="Intersecciones con el eje x")
plt.legend()
plt.grid()
plt.show()

for i, sol in enumerate(solutions):
    print(f"Solución aproximada {i+1}: x ≈ {sol:.4f}")

```

$d)$ $2senx=\sqrt{x+1}$

```python
import numpy as np
import matplotlib.pyplot as plt

def func_D(x):
    return 2 * np.sin(x) - np.sqrt(x + 1)

def dfunc_D(x):
    return 2 * np.cos(x) - 1 / (2 * np.sqrt(x + 1))

def newton_method(func, dfunc, x0, tolerance=0.0001, max_iter=100):
    x = x0
    for _ in range(max_iter):
        fx = func(x)
        if abs(fx) < tolerance:
            return x
        dfx = dfunc(x)
        if dfx == 0:
            break
        x = x - fx / dfx
    return None

x = np.linspace(0, 3, 100)
y = func_D(x)
plt.plot(x, y)
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title("f(x) = 2*sin(x) - sqrt(x + 1)")
plt.xlabel("x")
plt.ylabel("f(x)")

initial_guesses = [0.5, 1.0, 2.5]
intersections_x = []

solutions = []
for guess in initial_guesses:
    sol = newton_method(func_D, dfunc_D, guess)
    if sol is not None and all(abs(sol - s) > 0.0001 for s in solutions):
        solutions.append(sol)
        intersections_x.append(sol)

plt.scatter(intersections_x, [0]*len(intersections_x), color="red", zorder=5, label="Intersecciones con el eje x")
plt.legend()
plt.grid()
plt.show()

for i, sol in enumerate(solutions):
    print(f"Solución aproximada {i+1}: x ≈ {sol:.4f}")
```

Utilice el método de Newton para aproximar $\sqrt[3]{2}$ con un error menor a 0,00005. Utilice $x^n-k=0$

$Solucion: $

```python
def func(x):
    return x**3 - 2

def dfunc(x):
    return 3 * x**2

def newton_method(func, dfunc, x0, tolerance=0.00005):
    x = x0
    while abs(func(x)) > tolerance:
        x = x - func(x) / dfunc(x)
    return x

initial_guess = 1.5
solution = newton_method(func, dfunc, initial_guess)
print(f"Solución aproximada de [3]√{2}: x ≈ {solution:.5f}")
print("Profe no supimos poner la raiz cubica. Que pena")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función y sus derivadas
def f1(x):
    return x * (x + 1)

def f1_prime(x):
    return 2 * x + 1

def f1_double_prime(x):
    return 2 * np.ones_like(x)  # Array constante del mismo tamaño que x

# Valores para graficar
x = np.linspace(-3, 3, 400)
y = f1(x)
y_prime = f1_prime(x)
y_double_prime = f1_double_prime(x)

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = x(x + 1)", color="blue")
plt.plot(x, y_prime, label="f'(x) = 2x + 1", linestyle="--", color="orange")
plt.plot(x, y_double_prime, label="f''(x) = 2", linestyle=":", color="green")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([-1, 0], [0, 0], color="red", zorder=5, label="Intersecciones con el eje x")  # puntos de intersección
plt.title("Análisis de la función f(x) = x(x + 1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

```
